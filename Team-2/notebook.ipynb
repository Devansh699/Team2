```python
import pandas as pd
import numpy as np
import sklearn as sk
import os
import nltk
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
movies = movies.merge(credits,on = 'title')
```

```python
movies.head(1)
```

```python
movies = movies[['id', 'title', 'keywords','overview','original_language','tagline','release_date','cast','crew','genres']]
```

```python
movies.head()
movies.dtypes
```

```python
movies.dropna(inplace=True)
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
def convert(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
            L.append(i['name'])
    return L 
```

```python
movies['keywords'] = movies['keywords'].apply(convert)



```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else: 
            break
    return L 
```

```python
movies.head()
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['cast'].head(1)
```

```python

movies['release_date'] = movies['release_date'].apply(lambda x: x[0:4])
```

```python
def director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
def collapse (L):
    L1=[]
    for i in L:
        L1.append(i.replace(" ",""))
    return L1

     
```

```python
movies.head(100)
```

```python
movies['crew'] = movies['crew'].apply(collapse)
movies['cast'] = movies['cast'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.tail()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies['original_language']=movies['original_language'].apply(lambda x:x.split())
```

```python
movies.head(5)
```

```python
movies['tags'] = movies['keywords'] + movies['overview'] + movies['cast'] + movies['crew'] + movies['genres'] + movies['original_language']
```

```python
movies.head()
```

```python
new_movies = movies.drop(columns=['overview' , 'genres' , 'keywords' , 'cast' , 'crew','release_date','tagline','original_language'])
#new.head()
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x: " ".join(x))
```

```python
def stem(tag):
    y = []
    
    for i in tag.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
new_movies.head()
```

```python
vectors = cv.fit_transform(new_movies['tags']).toarray()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
dewsis = cosine_similarity(vectors)
```

```python
sorted(dewsis[0])
```

```python
def recommend(movie):
    index = new_movies[new_movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(dewsis[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:11]:
        print(new_movies.iloc[i[0]].title)
```

```python
recommend('Batman v Superman: Dawn of Justice')
```

```python
recommend("The Avengers")
```
